# Grouping and Aggregation

```python
# Simplify categories by taking only the first category
df['main_category'] = df['categories'].str.split(',').str[0]

# Median rating by brand and main category
median_ratings = df.groupby(['brand', 'main_category'])['reviews.rating'].median()
print(median_ratings)

# Optional: Average helpful votes by brand and main category
helpful_votes = df.groupby(['brand', 'main_category'])['reviews.numHelpful'].sum()
print(helpful_votes)
```

# Merging DataFrames

```python

# Load datasets
df1 = pd.read_csv(r"..\data\Kaggle_dataset\amazon_review.csv", encoding='latin-1')
df2 = pd.read_csv(r"..\data\Kaggle_dataset\AmazonProductReviewsDataset.csv", encoding='latin-1')

# Standardize column names (lowercase, strip spaces)
df1.columns = df1.columns.str.strip().str.lower()
df2.columns = df2.columns.str.strip().str.lower()

# Align column names: rename 'overall' -> 'reviews.rating' if exists
if 'overall' in df1.columns:
    df1.rename(columns={'overall': 'reviews.rating'}, inplace=True)
if 'overall' in df2.columns:
    df2.rename(columns={'overall': 'reviews.rating'}, inplace=True)

# Concatenate datasets row-wise
df = pd.concat([df1, df2], ignore_index=True)

# Remove duplicate column names (keep first occurrence)
if df.columns.duplicated().any():
    df = df.loc[:, ~df.columns.duplicated()]

# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Final check
print("Final shape:", df.shape)
print(df.columns.tolist())   # Show cleaned column list
print(df.head())
```

# Concatenating DataFrame

```python

df_combined = pd.concat([df, df2], axis=0, ignore_index=True)

print("Shape after concatenation:", df_combined.shape)
print(df_combined.head())
```

# Text Cleaning (removing unnecessary emojis, symbols etc)

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = re.sub(r"http\S+", "", text)        # remove URLs
    text = re.sub(r"&[a-z]+;", "", text)       # remove HTML entities
    text = re.sub(r"[^A-Za-z0-9\s]", "", text) # remove special characters
    text = text.lower().strip()
    return text

df['reviews.text'] = df['reviews.text'].apply(clean_text)
df['summary'] = df['summary'].apply(clean_text)
```
