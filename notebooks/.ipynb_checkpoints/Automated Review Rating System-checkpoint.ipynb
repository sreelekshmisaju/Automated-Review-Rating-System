# Importing Libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

# Load Data

```python
df = pd.read_csv(r"..\data\Kaggle_dataset\AmazonProductReviewsDataset.csv", encoding='latin-1')

print("Amazon dataset shape:", df.shape)
```

# View Data

```python
# Amazon dataset
print("\nAmazon Dataset:")
df.head()
df.tail()
df.sample(5)
df.info()
df.describe()

```

# Rename columns

```python

df.rename(columns={'reviews.text': 'Review', 'reviews.rating': 'Rating', 'asins': 'ProductDetails' }, inplace=True)
```

#  Handle Missing Data

```python
# Drop rows where Review or Rating is missing
df = df.dropna(subset=['Review', 'Rating'])

# Convert Rating to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Fill any remaining missing Rating with median
df['Rating'] = df['Rating'].fillna(df['Rating'].median())

# Remove rows where Review is empty, only spaces, or just a dot
df = df[~df['Review'].str.strip().isin(['', '.'])]

# Reset index after cleaning
df.reset_index(drop=True, inplace=True)

print("After handling missing and empty data:")
df.info()
df.head()
```

#  Remove Duplicates

```python

df = df.drop_duplicates(subset=['ProductDetails','Rating','Review'])
print("After removing duplicates:")
df.shape
df.head()
```

```python
#Remove exact duplicates (same review + rating)
df = df.drop_duplicates(subset=['Review', 'Rating'])

# Remove conflicting reviews (same text, different ratings)
conflict_idx = df[df.duplicated(subset=['Review'], keep=False) & 
                         df.duplicated(subset=['Review', 'Rating'], keep=False) == False].index
df.drop(index=conflict_idx, inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

print("After removing duplicates and conflicts:", df.shape)

```

#  Selecting Columns

```python
df = df[['ProductDetails','Rating','Review']]

print("Final dataset columns:", df.columns)

print("Final dataset shape:", df.shape)
```

#  Filtering Rows

```python
df_filtered = df[(df['Rating'] >= 1) & (df['Rating'] <= 5)]

print("Filtered dataset shape:", df_filtered.shape)
df_filtered.head()

```

#  Grouping and Aggregation

```python
df.groupby('ProductDetails')['Rating'].mean()
df.groupby('ProductDetails')['Review'].count()
df.groupby('ProductDetails')['Rating'].agg(['min', 'max'])
df.groupby('ProductDetails').agg({
    'Rating': ['mean', 'min', 'max', 'count'],  # multiple stats on Rating
    'Review': 'count'                           # count of reviews
})
```

#  Cleaning text - remove emojis, special characters, symbols

```python
def clean_text(text):
    text = re.sub(r'[^A-Za-z0-9\s]+', '', str(text))  # Remove special characters/emojis
    text = re.sub(r'\s+', ' ', text).strip()          # Remove extra spaces
    return text

df_filtered['Review'] = df_filtered['Review'].apply(clean_text)

```

# Remove Very Short Reviews

```python
# Remove reviews with fewer than 10 characters
df_filtered = df_filtered[df_filtered['Review'].str.len() >= 10]

# Reset index
df_filtered.reset_index(drop=True, inplace=True)

print("After removing very short reviews:", df.shape)
```

# Load the new dataset

```python
df1 = pd.read_csv(r"..\data\Kaggle_dataset\Womens Clothing E-Commerce Reviews.csv", encoding="latin-1")
```

# Rename columns to match your cleaned dataset

```python

df1.rename(columns={
    "Class Name": "ProductDetails",
    "Rating": "Rating",
    "Review Text": "Review"
}, inplace=True)


# Select the same columns as df_filtered
df1 = df1[["ProductDetails", "Rating", "Review"]]

```

# Concatenate with  cleaned dataset

```python

df_final = pd.concat([df_filtered, df1], axis=0).reset_index(drop=True)
```

```python
# Replace string 'nan' (and similar variants) with actual NaN
df_final['Review'] = df_final['Review'].replace(['nan', 'NaN', 'NAN', 'None'], np.nan)

# Drop rows where Review is NaN, empty, or just spaces/dots
df_final = df_final.dropna(subset=['Review'])
df_final = df_final[~df_final['Review'].str.strip().isin(['', '.'])]

# Reset index after cleaning
df_final.reset_index(drop=True, inplace=True)

```


# View and Inspect of Final merged Data


```python
print("Final merged dataset shape:", df_final.shape)
df_final.head()
df_final.tail()
```

# Sorting Data

```python
df_final_sorted = df_final.sort_values(by='Rating', ascending=False).reset_index(drop=True)
df_final_sorted.head(10)
```

```python
# Count of reviews per product class
class_distribution = df_final_sorted['ProductDetails'].value_counts()
print("Class distribution:\n", class_distribution)
```

```python
rating_distribution = df_final_sorted['Rating'].value_counts().sort_index()
print("Rating distribution:\n", rating_distribution)
```

#  Save Cleaned Dataset

```python
df_final_sorted.to_csv(r"..\data\cleaned_product_reviews5.csv", index=False)

df_final_sorted.to_excel(r"..\data\cleaned_product_reviews4.xlsx", index=False)

print("Files saved successfully")
```

# Data Visualization

# Rating Distribution (Bar Chart)

```python
sns.countplot(x='Rating', data=df_final_sorted, color='skyblue')  
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()
```

# Review Length Distribution (Histogram)

```python

df_final_sorted['ReviewLength'] = df_final_sorted['Review'].str.len()


sns.histplot(df_final_sorted['ReviewLength'], bins=10, kde=True, color='skyblue')
plt.title('Review Length Distribution')
plt.xlabel('Number of Characters in Review')
plt.ylabel('Frequency')
plt.show()
```

# Boxplot of Review Length per Rating

```python
sns.boxplot(x='Rating', y='ReviewLength', data=df_final_sorted, color='green')
plt.title('Boxplot of Review Length per Rating')
plt.xlabel('Rating')
plt.ylabel('Review Length')
plt.show()
```

# Violin Plot (Density + Distribution of Review Length per Rating)

```python
sns.violinplot(x='Rating', y='ReviewLength', data=df_final_sorted, color='orange')
plt.title('Violin Plot of Review Length per Rating')
plt.xlabel('Rating')
plt.ylabel('Review Length')
plt.show()
```

# Creating  Balanced Dataset 

```python
min_count = 2000
balanced_df_list = []

for rating in df_final_sorted['Rating'].unique():
    df_rating = df_final_sorted[df_final_sorted['Rating'] == rating]
    
    if len(df_rating) < min_count:
        # Oversample minority class
        df_rating_balanced = df_rating.sample(min_count, replace=True, random_state=42)
    else:
        # Undersample majority class
        df_rating_balanced = df_rating.sample(min_count, replace=False, random_state=42)
    
    balanced_df_list.append(df_rating_balanced)

df_balanced = pd.concat(balanced_df_list).sample(frac=1, random_state=42).reset_index(drop=True)

print(df_balanced['Rating'].value_counts())
```
